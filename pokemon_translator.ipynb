```python
import pandas as pd
from googletrans import Translator

def is_pokemon(text):
    key_words = ['pokemongo', 'pokecoins']
    temp_text = " ".join(filter(lambda x:x[0]!='@', text.split()))
    new_text = temp_text.lower().replace(' ', '').replace('Ã©', 'e')
    for key in key_words:
        if key in new_text:
            return True
    return False

def translate_text2(row):
    if row['lang'] != 'en':
        try:
            trans = Translator().translate(row['text'].replace('\n', '.'), dest='en')
            translated_text = trans.text
            row['text'], row['lang'] = translated_text, 'en'
        except:
            with open('FailedTranslations.txt', 'a') as logFile:
                logFile.write(row['text'].replace('\n', '.') + '\n--------------------\n')
    return row

df = pd.read_csv('SparkAllTweets.csv')
df = df[df.apply(lambda x: is_pokemon(x['text']), axis=1)]
print(df.count())
df = df.apply(translate_text2, axis=1)
print(df.count())
df.to_csv('translated_pokemon_tweets.csv')

df = pd.read_csv('SparkAllRetweets.csv')
df = df[df.apply(lambda x: is_pokemon(x['text']), axis=1)]
print(df.count())
df = df.apply(translate_text2, axis=1)
print(df.count())
df.to_csv('translated_pokemon_retweets.csv')
```
